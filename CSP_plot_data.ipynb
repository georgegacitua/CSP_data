# Read the Data

```python
import datetime
import numpy as np
import matplotlib.pyplot as plt

#Data
data_36mw = np.genfromtxt("results_36mw.csv", delimiter = ',', skip_header = 1)[:, 2:]
data_50mw = np.genfromtxt("results_50mw.csv", delimiter = ',', skip_header = 1)[:, 2:]
data_72mw = np.genfromtxt("results_72mw.csv", delimiter = ',', skip_header = 1)[:, 2:]
data_110mw = np.genfromtxt("results_110mw.csv", delimiter = ',', skip_header = 1)[:, 2:]

base = datetime.date(2020, 1, 1)
date_list = [base + datetime.timedelta(days=x) for x in range(365)]
```

# Daily Power Output
## Mean Output

```python
#Get PC power output mean and max
pc_output_36mw = data_36mw[:, 2].reshape((365, 24)).T
mean_output_36mw = np.mean(pc_output_36mw, axis= 0)
max_output_36mw = np.max(pc_output_36mw, axis= 0)
pc_output_50mw = data_50mw[:, 2].reshape((365, 24)).T
mean_output_50mw = np.mean(pc_output_50mw, axis= 0)
max_output_50mw = np.max(pc_output_50mw, axis= 0)
pc_output_72mw = data_72mw[:, 2].reshape((365, 24)).T
mean_output_72mw = np.mean(pc_output_72mw, axis= 0)
max_output_72mw = np.max(pc_output_72mw, axis= 0)
pc_output_110mw = data_110mw[:, 2].reshape((365, 24)).T
mean_output_110mw = np.mean(pc_output_110mw, axis= 0)
max_output_110mw = np.max(pc_output_110mw, axis= 0)

#Graph
plt.figure(figsize=(12, 6)) #Image Size
plt.plot(date_list, mean_output_110mw, label= '110 MW')
plt.plot(date_list, mean_output_72mw, label= '72 MW')
plt.plot(date_list, mean_output_50mw, label= '50 MW')
plt.plot(date_list, mean_output_36mw, label= '36 MW')
plt.xlabel(r'Days', fontsize=16)
plt.ylabel(r'Electrical power output [MWe]', fontsize=16)
plt.title("Daily mean PC power output ", fontsize=18)
plt.xlim(datetime.date(2020, 1, 1), datetime.date(2020, 12, 31))
plt.legend()
plt.grid()

#Save figure
plt.savefig('daily_power_output.png', format='png', bbox_inches='tight')
plt.show()
```

## Max Output

```python
#Graph
plt.figure(figsize=(12, 6)) #Image Size
plt.plot(date_list, max_output_110mw, label= '110 MW')
plt.plot(date_list, max_output_72mw, label= '72 MW')
plt.plot(date_list, max_output_50mw, label= '50 MW')
plt.plot(date_list, max_output_36mw, label= '36 MW')
plt.xlabel(r'Days', fontsize=16)
plt.ylabel(r'Electrical power output [MWe]', fontsize=16)
plt.title("Daily maximum PC power output ", fontsize=18)
plt.xlim(datetime.date(2020, 1, 1), datetime.date(2020, 12, 31))
plt.legend()
plt.grid()

#Save figure
plt.savefig('daily_max_power_output.png', format='png', bbox_inches='tight')
plt.show()
```

# PC Efficiency

```python
#Get PC power output mean and max
efficiency_36mw = data_36mw[:, 1].reshape((365, 24)).T
mean_efficiency_36mw = np.mean(efficiency_36mw, axis= 0)
efficiency_50mw = data_50mw[:, 1].reshape((365, 24)).T
mean_efficiency_50mw = np.mean(efficiency_50mw, axis= 0)
efficiency_72mw = data_72mw[:, 1].reshape((365, 24)).T
mean_efficiency_72mw = np.mean(efficiency_72mw, axis= 0)
efficiency_110mw = data_110mw[:, 1].reshape((365, 24)).T
mean_efficiency_110mw = np.mean(efficiency_110mw, axis= 0)

#Graph
plt.figure(figsize=(12, 6)) #Image Size
plt.plot(date_list, mean_efficiency_110mw, label= '110 MW')
plt.plot(date_list, mean_efficiency_72mw, label= '72 MW')
plt.plot(date_list, mean_efficiency_50mw, label= '50 MW')
plt.plot(date_list, mean_efficiency_36mw, label= '36 MW')
plt.xlabel(r'Days', fontsize=16)
plt.ylabel(r'PC efficiency', fontsize=16)
plt.title("Daily mean PC efficiency", fontsize=18)
plt.xlim(datetime.date(2020, 1, 1), datetime.date(2020, 12, 31))
plt.legend()
plt.grid()

#Save figure
plt.savefig('daily_efficiency.png', format='png', bbox_inches='tight')
plt.show()
```

# Daily Water Consumption (makeup + cooling)

```python
#Get volume of water consumed
water_36mw = data_36mw[:, 3].reshape((365, 24)).T * 3600 / 1000
daily_water_con_36mw = np.sum(water_36mw, axis= 0)
water_50mw = data_50mw[:, 3].reshape((365, 24)).T * 3600 / 1000
daily_water_con_50mw = np.sum(water_50mw, axis= 0)
water_72mw = data_72mw[:, 3].reshape((365, 24)).T * 3600 / 1000
daily_water_con_72mw = np.sum(water_72mw, axis= 0)
water_110mw = data_110mw[:, 3].reshape((365, 24)).T * 3600 / 1000
daily_water_con_110mw = np.sum(water_110mw, axis= 0)

#Graph
plt.figure(figsize=(12, 6)) #Image Size
plt.plot(date_list, daily_water_con_110mw, label= '110 MW')
plt.plot(date_list, daily_water_con_72mw, label= '72 MW')
plt.plot(date_list, daily_water_con_50mw, label= '50 MW')
plt.plot(date_list, daily_water_con_36mw, label= '36 MW')
plt.xlabel(r'Days', fontsize=16)
plt.ylabel(r'PC water consumption [$m^3$]', fontsize=16)
plt.title("Daily PC water consumption ", fontsize=18)
plt.xlim(datetime.date(2020, 1, 1), datetime.date(2020, 12, 31))
plt.legend()
plt.grid()

#Save figure
plt.savefig('daily_water_consumption.png', format='png', bbox_inches='tight')
plt.show()
```
