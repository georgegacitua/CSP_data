# READ DATA

```python
#Libraries
import datetime
import numpy as np
import matplotlib.pyplot as plt

#Data
dry_data = np.genfromtxt("cerro_dominador_dry.csv", delimiter = ',', skip_header = 1)[:, 1:]
wet_data = np.genfromtxt("cerro_dominador_wet.csv", delimiter = ',', skip_header = 1)[:, 1:]

base = datetime.date(2020, 1, 1)
date_list = [base + datetime.timedelta(days=x) for x in range(365)]
```

# Daily efficiency

```python
#Get mean efficiency
dry_efficiency = dry_data[:, 1].reshape((365, 24)).T
mean_dry_efficiency = np.mean(dry_efficiency, axis= 0)
wet_efficiency = wet_data[:, 1].reshape((365, 24)).T
mean_wet_efficiency = np.mean(wet_efficiency, axis= 0)

#Graph
plt.figure(figsize=(12, 6)) #Image Size
plt.plot(date_list, mean_dry_efficiency, label= 'Dry PC')
plt.plot(date_list, mean_wet_efficiency, label= 'Wet PC')
plt.plot([datetime.date(2020, 1, 1), datetime.date(2020, 12, 31)],[.4, .4],'--', label='Nominal Efficiency')
#plt.xlabel(r'Days', fontsize=16)
plt.ylabel(r'Rankine Cycle Efficiency', fontsize=16)
plt.title("Daily Mean Efficiency", fontsize=18)
plt.xlim(datetime.date(2020, 1, 1), datetime.date(2020, 12, 31))
plt.ylim(0, .5)
plt.legend()
plt.grid()

#Save figure
#plt.savefig('daily_mean_efficiency.png', format='png', bbox_inches='tight')
plt.show()
```

```python
labels = ['Dry PC', 'Wet PC']
power = [86.5, 91.2]


#Graph
plt.figure(figsize=(8, 6)) #Image Size
plt.bar(labels, power)
plt.xlabel(r'Cooling Method', fontsize=16)
plt.ylabel(r'Percentage [$\%$]', fontsize=16)
plt.ylim(80, 100)
plt.title("Capacity Factor", fontsize=18)

plt.show()
```

# Daily Power Output

```python
#Get Output Gross
dry_gross = dry_data[:, 2].reshape((365, 24)).T
mean_dry_gross = np.mean(dry_gross, axis= 0)
wet_gross = wet_data[:, 2].reshape((365, 24)).T
mean_wet_gross = np.mean(wet_gross, axis= 0)
#Get power to grid
dry_grid = dry_data[:, 4].reshape((365, 24)).T
mean_dry_grid = np.mean(dry_grid, axis= 0)
wet_grid = wet_data[:, 4].reshape((365, 24)).T
mean_wet_grid = np.mean(wet_grid, axis= 0)

#Graph
plt.figure(figsize=(12, 6)) #Image Size
plt.plot(date_list, mean_dry_gross, label= 'Dry PC gross output')
plt.plot(date_list, mean_wet_gross, label= 'Wet PC gross output')
plt.plot(date_list, mean_dry_grid, label= 'Dry PC power to grid')
plt.plot(date_list, mean_wet_grid, label= 'Wet PC power to grid')
#plt.xlabel(r'Days', fontsize=16)
plt.ylabel(r'Electric Power [MW]', fontsize=16)
plt.title("PC Electric Power delivered", fontsize=18)
plt.xlim(datetime.date(2020, 1, 1), datetime.date(2020, 12, 31))
plt.ylim(40, 120)
plt.legend()
plt.grid()

#Save figure
#plt.savefig('daily_power.png', format='png', bbox_inches='tight')
plt.show()
```

# Water Usage

```python
#Water consumption
dry_w_usage = dry_data[:, 3].reshape((365, 24)).T
total_dry_w_usage = np.mean(dry_w_usage, axis= 0)
wet_w_usage = wet_data[:, 3].reshape((365, 24)).T
total_wet_w_usage = np.mean(wet_w_usage, axis= 0)

#Graph
plt.figure(figsize=(12, 6)) #Image Size
plt.plot(date_list, total_dry_w_usage, label= 'Dry PC water consumption')
plt.plot(date_list, total_wet_w_usage, label= 'Wet PC water consumption')
#plt.xlabel(r'Days', fontsize=16)
plt.ylabel(r'Water Flow [kg/s]', fontsize=16)
plt.title("PC water consumption makeup + cooling", fontsize=18)
plt.xlim(datetime.date(2020, 1, 1), datetime.date(2020, 12, 31))
plt.legend()
plt.grid()

#Save figure
#plt.savefig('daily_water_consumption.png', format='png', bbox_inches='tight')
plt.show()
```

# Parasitic Losses

```python
#Get parasitic loss
dry_parasitic = dry_data[:, 5].reshape((365, 24)).T
mean_dry_parasitic = np.mean(dry_parasitic, axis= 0)
wet_parasitic = wet_data[:, 5].reshape((365, 24)).T
mean_wet_parasitic = np.mean(wet_parasitic, axis= 0)

#Graph
plt.figure(figsize=(12, 6)) #Image Size
plt.plot(date_list, mean_dry_parasitic, label= 'Dry PC parasitic power operation')
plt.plot(date_list, mean_wet_parasitic, label= 'Wet PC parasitic power operation')
#plt.xlabel(r'Days', fontsize=16)
plt.ylabel(r'Electric Power [MW]', fontsize=16)
plt.title("Condenser Power Parasitic Operation", fontsize=18)
plt.xlim(datetime.date(2020, 1, 1), datetime.date(2020, 12, 31))
plt.legend()
plt.grid()

#Save figure
#plt.savefig('daily_parasitic_operation.png', format='png', bbox_inches='tight')
plt.show()



```
